### Exploratory Data Analysis (EDA)
   - We explore the dataset using various descriptive statistics and visualizations.  
   -  Compute summary statistics like mean, median, and standard deviation.
- Identify patterns, correlations, and distributions in the data. 

```python
# Summary statistics
print(df.describe())
```

```python
# Visualize: Gender distribution
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette='viridis')
plt.title('Gender Distribution')
plt.ylabel('Count')
plt.xlabel('Gender')
plt.show()
```

```python
#Checking the distribution of height
# Histogram of height
plt.figure(figsize=(8, 5))
sns.histplot(df['Height'], bins=30, kde=True, color='blue')
plt.title("Distribution of Height")
plt.xlabel("Height (inches)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Checking the distribution of  weight
# Histogram of weight
plt.figure(figsize=(10, 5))
sns.histplot(df['Weight'], bins=30, kde=True, color='red')
plt.title("Distribution of Weight")
plt.xlabel("Weight (pounds)")
plt.ylabel("Frequency")
plt.show()
```

```python
#group data by
# Group data by Gender and calculate mean height and weight
gender_means = df.groupby('Gender')[['Height', 'Weight']].mean()  #group the data 
# Display mean values
print(gender_means)
```

```python
# Group the data by gender
grouped_data = df.groupby('Gender')
# Calculate the mean height and weight for each gender
avg_values = grouped_data[['Height', 'Weight']].mean()
print("Average Height and Weight by Gender:")
print(avg_values)
```

```python
# calculate the media height and weight
median_values=grouped_data[['Height', 'Weight', ]].median()
print('median Height and weight by Gender:')
print(median_values)
```

```python
#calculate standard deviation of Height and weight
std_values=grouped_data[['Height','Weight']].std()
print('standard deviation of Height and Weight:')
print(std_values)
```

```python
#calculate the variance of Height and Weight
variance_values=grouped_data[['Height', 'Weight']].var()
print('variance of Height and Weight:')
print(variance_values)
```

```python
#Calculate the min and max height and weight
min_values=grouped_data[['Height', 'Weight']].min()
max_values=grouped_data[['Height', 'Weight']].max()
print('min values of Height and Weight:')
print(min_values)
print('max values of Height and Weight:')
print(max_values)
```

```python
#calculate the range of Height and Weight
range_values=max_values-min_values
print('range of Height and Weight:')
print(range_values)
```

```python

## checking the  Distribution of Height gy gender
# Histograms for height and weight by gender
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='Height', hue='Gender', kde=True, bins=20, alpha=0.5)
plt.title('Distribution of Height by Gender')
plt.xlabel('Height')
plt.ylabel('Frequency')
plt.show()
```

```python
## checking the  Distribution of weight gy gender
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='Weight', hue='Gender', kde=True, bins=20, alpha=0.5)
plt.title('Distribution of Weight by Gender')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate IQR for height
q1_height = df['Height'].quantile(0.25)
q3_height = df['Height'].quantile(0.75)
iqr_height = q3_height - q1_height
lower_bound_height = q1_height - 1.5 * iqr_height
upper_bound_height = q3_height + 1.5 * iqr_height
print(q1_height)
print(q3_height)
print(lower_bound_height)
print(upper_bound_height)
```

```python
# Calculate IQR for weight
q1_weight = df['Weight'].quantile(0.25)
q3_weight = df['Weight'].quantile(0.75)
iqr_weight = q3_height - q1_height
lower_bound_weight = q1_height - 1.5 * iqr_height
upper_bound_weight = q3_height + 1.5 * iqr_height
print(q1_weight)
print(q3_weight)
print(lower_bound_height)
print(upper_bound_height)
```

```python
# Detect outliers

outliers_height = df[(df['Height'] < lower_bound_height) | (df['Height'] > upper_bound_height)]
outliers_weight = df[(df['Weight'] < lower_bound_weight) | (df['Weight'] > upper_bound_weight)]
print(f"Outliers in Height: {len(outliers_height)}")
print(f"Outliers in Weight: {len(outliers_weight)}")
```

```python
# Box plot
plt.figure(figsize=(10, 5))

# Subplot 1: Height Distribution
plt.subplot(1, 2, 1)
sns.boxplot(y=df["Height"], color='skyblue')
plt.title("Box Plot: Height Distribution")
plt.ylabel("Height (inches)")

# Subplot 2: Weight Distribution
plt.subplot(1, 2, 2)
sns.boxplot(y=df["Weight"], color='lightgreen')
plt.title("Box Plot: Weight Distribution")
plt.ylabel("Weight (pounds)")

# Show plot
plt.tight_layout()
plt.show()
```

#### Box Plot: Distribution of Height & Weight
- The box represents the interquartile range (IQR) (middle 50% of data).
- The line inside the box represents the median (middle value).
- Whiskers extend to the smallest/largest non-outlier values.
- Dots outside the whiskers indicate outliers.

### Analyzing Correlation
   - The correlation coefficient helps quantify the relationship between height and weight.

```python
# Compute correlation
correlation = df["Height"].corr(df["Weight"])
print(f"Correlation between Height and Weight: {correlation:.2f}")
```

```python
# Scatter plot
plt.figure(figsize=(10, 5))
sns.scatterplot(x=df["Height"], y=df["Weight"], alpha=0.5, color='blue')

# Labels and title
plt.title("Scatter Plot: Height vs. Weight")
plt.xlabel("Height (inches)")
plt.ylabel("Weight (pounds)")

# Show plot
plt.show()
```

#####  Scatter Plot: Height vs. Weight
- Each dot represents an individual data point, mapping Height (x-axis) against Weight (y-axis).
- If the dots trend upward, it shows a positive correlation (i.e., taller individuals tend to have higher weight).
- If the points are widely scattered, the relationship may be weak

### Insights and Conclusion
- The histograms provide insights into the distribution of height and weight across the dataset.
- The scatter plot shows a clear positive correlation between height and weight, meaning taller individuals tend to weigh more.
- The correlation coefficient quantifies the strength of this relationship.
